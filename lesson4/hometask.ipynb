```python
documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}

def name_by_document_number():
    number = input('Введите номер документа: ')
    for el in documents:
        if number in el.values():
            print('Владелец документа:', el['name'])
            break
        elif el == documents[-1]:
            print('Документ не найден в базе')

def directories_by_document_number():
    number = input('Введите номер документа: ')
    docs = list(directories.items())
    for directorie in docs:
        if number in directorie[1]:
            print('Документ хранится на полке:', directorie[0])
            break
        elif directorie == docs[-1]:
            print('Документ не найден в базе')
            
def show_all_data():
    data = []
    for document in documents:
        document_copy = document.copy()
        num = document['number']
        docs = list(directories.items())
        for directorie in docs:
            if num in directorie[1]:
                document_copy.setdefault('directorie', directorie[0])
        data.append(document_copy)
        print('№:', document_copy['number'] + ',', 'тип:', document_copy['type'] + ',',\
              'владелец:', document_copy['name'] + ',', 'полка храненения:', document_copy['directorie'])
    return data
def add_directorie():
    number = input('Введите номер полки:')
    directorie = []
    keys = list(directories.keys())
    if number in list(directories.keys()):
        print('Такая полка уже существует. Текущий перечень полок:', ', '.join(keys))
    else:
        directories.setdefault(number, directorie)
        keys.append(number)
        print('Полка добавлена. Текущий перечень полок:', ', '.join(keys))
    return directories

def del_directorie():
    number = input('Введите номер полки:')
    keys = list(directories.keys())
    if directories[number] == []:
        del(directories[number])
        keys = list(directories.keys())
        print('Полка удалена. Текущий перечень полок:', ', '.join(keys))
    else:
        print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок:', ', '.join(keys))
    return directories

def create_document(number, typ, name, direct):
    documents.append({'type': typ, 'number': number, 'name': name})
    directories[direct].append(number)

def addit(number, typ, name, direct):
    if direct in directories.keys():
        create_document(number, typ, name, direct)
        print('Документ добавлен. Текущий список документов:')
    else:
        print('Такой полки не существует. Добавьте полку командой ads.')
    show_all_data()

def add_document():
    number = input('Введите номер документа: ')
    typ = input('Введите тип документа: ')
    name = input('Введите владельца документа: ')
    direct = input('Введите полку для хранения: ')
    addit(number, typ, name, direct)

def all_numbers_in_directories():
    directs = list(directories.values())
    all_num = []
    for direct in directs:
        for num in direct:
            all_num.append(num)
    return all_num

def removment(number, docs):
    for el in documents:
        if number in el.values():
            documents.remove(el)
    for directorie in docs:
        if number in directorie:
            directorie.remove(number)


def delitements(number):
    docs = list(directories.values())
    if number in all_numbers_in_directories():
        removment(number, docs)
        print('Документ удалён.')
        print('Текущий список документов:')
        show_all_data()
    else:
        print('Документ не найден в базе')
        print('Текущий список документов:')
        show_all_data()

def delite_document():
    number = input('Введите номер документа: ')
    delitements(number)
    
def move_document():
    number = input('Введите номер документа: ')
    direct = input('Введите номер полки: ')
    if number in all_numbers_in_directories() and direct in directories.keys():
        docs = list(directories.values())
        for el in documents:
            if number in el.values():
                typ = el['type']
                name = el['name']
        removment(number, docs)
        create_document(number, typ, name, direct)
        print('Документ перемещён.')
        print('Текущий список документов:')
        show_all_data()
    elif number not in all_numbers_in_directories():
        print('Документ не найден в базе')
        print('Текущий список документов:')
        show_all_data()
    elif direct not in directories.keys():
        keys = list(directories.keys())
        print('Такой полки не существует. Текущий перечень полок:', ', '.join(keys))
        
        
def take_command(command): 
    if command == 'p':
        name_by_document_number()
    elif command == 's':
        directories_by_document_number()
    elif command == 'l':
        show_all_data()
    elif command == 'ads':
        add_directorie()
    elif command == 'ds':
        del_directorie()
    elif command == 'ad':
        add_document()
    elif command == 'd':
        delite_document()
    elif command == 'm':
        move_document()
```

```python
while command != 'q':
    command = input('Введите команду: ')
    take_command(command)
```
